```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string
```

```python
spark.conf.set("spark.sql.parquet.datetimeRebaseModeInRead", "LEGACY")
```

```python
df = spark.sql("SELECT * FROM SephoraLake.sentiment_analysis_comment").toPandas()
print(df)
```

```python
nltk.download('stopwords')
nltk.download('punkt')
```

```python
def preprocess_text(text):
    # تحويل النص إلى حروف صغيرة
    text = text.lower()
    # إزالة الترقيم
    text = text.translate(str.maketrans('', '', string.punctuation))
    # إزالة الكلمات التوقف
    stop_words = set(stopwords.words('english'))
    words = word_tokenize(text)
    filtered_words = [word for word in words if word not in stop_words]
    return ' '.join(filtered_words)
```

```python
df['processed_text'] = df['review_text'].apply(preprocess_text)
```

```python
# Define the mapping
sentiment_mapping = {
    'Positive': 1.0,
    'Negative': 0.0,
    'neutral': 0.5
}

# Create a new column with descriptive labels
df['sentiment_label'] = df['sentiment'].map(sentiment_mapping)
df['sentiment_label'].fillna(5, inplace=True)

# Display the DataFrame
print(df)
df.head()
df['sentiment_label'].unique()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df['processed_text'], df['sentiment_label'], test_size=0.2, random_state=42)
```

```python
vectorizer = CountVectorizer()
X_train_vectors = vectorizer.fit_transform(X_train)
X_test_vectors = vectorizer.transform(X_test)
```

```python
model = LogisticRegression()
model.fit(X_train_vectors, y_train)
```

```python
# Predict and evaluate the model
y_pred = model.predict(X_test_vectors)
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print(f"Accuracy: {accuracy}")
print(f"Classification Report:\n{report}")
```
