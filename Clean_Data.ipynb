```python
# Welcome to your new notebook
# Type here in the cell editor to add code!
import pandas as pd
```

```python
File_Path = 'abfss://Sephora_@onelake.dfs.fabric.microsoft.com/SephoraLake.Lakehouse/Files/product_info.csv'
New_File_Path = 'abfss://Sephora_@onelake.dfs.fabric.microsoft.com/SephoraLake.Lakehouse/Files/Product_Info.csv'
df = pd.read_csv(File_Path)
df.head()
```

```python
df.duplicated()
```

```python
# Check for missing values
missing_values = df.isnull().sum()
missing_values
```

```python
# Drop columns with excessive missing values
columns_to_drop = ['variation_desc', 'value_price_usd', 'sale_price_usd']
df = df.drop(columns=columns_to_drop)

# Fill missing values for numeric columns
df['rating'] = df['rating'].fillna(df['rating'].mean())
df['reviews'] = df['reviews'].fillna(df['reviews'].mean())
df['child_max_price'] = df['child_max_price'].fillna(df['child_max_price'].mean())
df['child_min_price'] = df['child_min_price'].fillna(df['child_min_price'].mean())

# Fill missing values for categorical columns
df['size'] = df['size'].fillna('Unknown')
df['variation_type'] = df['variation_type'].fillna('Unknown')
df['variation_value'] = df['variation_value'].fillna('Unknown')
df['ingredients'] = df['ingredients'].fillna('Unknown')
df['highlights'] = df['highlights'].fillna('Unknown')
df['secondary_category'] = df['secondary_category'].fillna('Unknown')
df['tertiary_category'] = df['tertiary_category'].fillna('Unknown')

# Check if there are any missing values left
df.isnull().sum()
```
